```python
def cropImage(imgPath):
    image = cv2.imread(imgPath)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        colorImg = image[y:y+h, x:x+w]
        if (len(faces)>0):
            return (colorImg)
        
    
```

```python
path_to_data = "./rawDataset/"
path_to_crop = "./Dataset/"
```

```python
import os
img_dirs = []
for entry in os.scandir(path_to_data):
    if entry.is_dir():
        img_dirs.append(entry.path)
        
img_dirs
```

```python
os.mkdir(path_to_crop)
```

```python
cropped_image_dirs = []
char_file_names_dict = {}

for img_dir in img_dirs:
    count = 1
    char_name = img_dir.split('/')[-1]
    print(char_name)
    
    char_file_names_dict[char_name] = []
    
    for entry in os.scandir(img_dir):
        roi_color = cropImage(entry.path)
        if roi_color is not None:
            cropped_folder = path_to_crop + char_name
            if not os.path.exists(cropped_folder):
                os.makedirs(cropped_folder)
                cropped_image_dirs.append(cropped_folder)
                print("Generating cropped images in folder: ",cropped_folder)
                
            cropped_file_name = char_name + str(count) + ".png"
            cropped_file_path = cropped_folder + "/" + cropped_file_name 
            
            cv2.imwrite(cropped_file_path, roi_color)
            char_file_names_dict[char_name].append(cropped_file_path)
            count += 1
```

```python
final_img_dirs =[]
for entry in os.scandir(path_to_crop):
    if entry.is_dir():
        final_img_dirs.append(entry.path)
final_img_dirs 
```

```python
final_count = 0
for final_img_dir in final_img_dirs:
    
    for entry in os.scandir(final_img_dir):
        final_count += 1
    print(final_count, final_img_dir.split('/')[-1])
    final_count = 0
    
```
