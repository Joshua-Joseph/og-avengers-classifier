```python
import numpy as np
import pywt
import cv2 
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
import os
```

```python
def w2d(img, mode='haar', level=1):
    imArray = img
    #Datatype conversions
    #convert to grayscale
    imArray = cv2.cvtColor( imArray,cv2.COLOR_RGB2GRAY )
    #convert to float
    imArray =  np.float32(imArray)   
    imArray /= 255;
    # compute coefficients 
    coeffs=pywt.wavedec2(imArray, mode, level=level)

    #Process Coefficients
    coeffs_H=list(coeffs)  
    coeffs_H[0] *= 0;  

    # reconstruction
    imArray_H=pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255;
    imArray_H =  np.uint8(imArray_H)

    return imArray_H
```

```python
final_img_dirs = ['./Dataset/thor',
 './Dataset/hulk',
 './Dataset/cap',
 './Dataset/widow',
 './Dataset/iron',
 './Dataset/hawk']
```

```python
char_file_names_dict = {}
for img_dir in final_img_dirs:
    char_name = img_dir.split('/')[-1]
    file_list = []
    for entry in os.scandir(img_dir):
        file_list.append(entry.path)
    char_file_names_dict[char_name] = file_list
char_file_names_dict
```

```python
class_dict = {}
count = 0
for char_name in char_file_names_dict.keys():
    class_dict[char_name] = count
    count = count + 1
class_dict
```

```python
#found .DS_STORE file in each directory
for char_name, training_files in char_file_names_dict.items():
    for training_image in training_files:
        img = cv2.imread(training_image)
       
        if(type(img) == type(None)):
            print(training_image, 'not found')
            
        else:
            height, width, channels = img.shape
            print(training_image, height, width, channels)
```

```python
X, y = [], []
for char_name, training_files in char_file_names_dict.items():
    for training_image in training_files:
        img = cv2.imread(training_image)
        if(type(img) == type(None)):
            pass
        else:
            scaled_raw_img = cv2.resize(img, (32, 32))
            img_har = w2d(img,'db1',5)
            scaled_img_har = cv2.resize(img_har, (32, 32))
            combined_img = np.vstack((scaled_raw_img.reshape(32*32*3,1),scaled_img_har.reshape(32*32,1)))
        X.append(combined_img)
        y.append(class_dict[char_name])
```

```python
y[0]
```

```python
X = np.array(X).reshape(len(X),4096).astype(float)
X.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.decomposition import PCA as RandomizedPCA
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
n_components = 150
pca = RandomizedPCA(n_components=n_components, whiten=True).fit(X_train)
print(pca.explained_variance_ratio_)
```

```python

X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
param_grid = {
         'C': [1e3, 5e3, 1e4, 5e4, 1e5],
          'gamma': [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.1],
          }
clf = GridSearchCV(SVC(kernel='rbf', class_weight='balanced'), param_grid) 
clf = clf.fit(X_train_pca, y_train)

print("Best estimator found by grid search:")
print(clf.best_estimator_)
```

```python
y_pred = clf.predict(X_test_pca)
acc = accuracy_score(y_test,y_pred)
print('Accuracy:',acc)
```

```python
cm = confusion_matrix(y_test, y_pred)
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
